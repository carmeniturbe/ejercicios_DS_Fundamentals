```python
#a. ¿Cuántos atletas que participaron en el deporte "Tug-Of-War" (tira y afloja) pertenecían en el equipo “Sweden” (sueco) en el conjunto de datos de muestra? R: 6 atletas
filter = (df['Sport'] == 'Tug-Of-War') & (df['Team'] == 'Sweden')
tug_of_war_sweden_filter = df[filter]
print(tug_of_war_sweden_filter)
print(len(tug_of_war_sweden_filter))
```

```python
#b. ¿Cuántas medallas de oro se le entregaron a los atletas que tenían más de 40 años? R: 66 Medallas fueron entregadas
filter = (df['Age'] > 40) & (df['Medal'] == 'Gold')
gold_medals_atletes_over_40 = df[filter]
print(gold_medals_atletes_over_40)
print(len(gold_medals_atletes_over_40))

```

```python
#Determinen la altura promedio de:
#a. Todas las filas incluidas. - R= La altura promedio es 175.38 cm
df['Height'].mean()
```

```python
#Determinen la altura promedio de:
#b. Todas las filas incluidas, pero agrupadas por Event (evento).
df.groupby('Event')['Height'].mean()
```
