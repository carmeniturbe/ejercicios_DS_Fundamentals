```python
#Tracen un histograma del porcentaje de dueños de perros.
porcentaje_perros = catsvdogs['Percentage of Dog Owners']
sns.histplot(porcentaje_perros, bins=10, kde=False)

plt.xlabel('Porcentaje de Dueños de Perros')
plt.ylabel('Frecuencia')
plt.title('Distribución del Porcentaje de Dueños de Perros')

plt.show()

# Lo más común es tener localidades que tengan entre 30% y 45% de dueños de gatos.
# Tiene una distribución similar a una con carga a la izquierda.
```

```python
#Tracen un histograma del porcentaje de dueños de gatos.
porcentaje_gatos = catsvdogs['Percentage of Cat Owners']
sns.histplot(porcentaje_gatos, bins=10, kde=False)

plt.xlabel('Porcentaje de Dueños de Gatos')
plt.ylabel('Frecuencia')
plt.title('Distribución del Porcentaje de Dueños de Gatos')

plt.show()

# Lo más común es tener localidades que tengan entre 25% y 35% de dueños de gatos.
# Tiene una distribución similar a la normal.
```

```python
#tracen estos dos histogramas en el mismo trazado. Un ejemplo de lo que debería verse se proporciona abajo:
sns.histplot(porcentaje_perros, bins=8, kde=False)
sns.histplot(porcentaje_gatos, bins=15, kde=False)
#ajustamos etiquestas
plt.xlabel('Percentage of Owners')
plt.ylabel('Frequency')
plt.title('Distribution of Cat and Dog Owners')

#legend
plt.legend()  #intente agregar una leyenda pero no funciono
```

```python
#Creen dos boxplots en mismo trazado: uno en la media del número de gatos por hogar y el otro de la media del número de perros.

# Combinamos las columnas de interés en una sola columna llamada "Type"
catsvdogs_melted = catsvdogs.melt(value_vars=['Mean Number of Dogs per household', 'Mean Number of Cats'], var_name='Type')

# Trama de boxplot con ambos boxplots
sns.catplot(data=catsvdogs_melted, x='Type', y='value', kind='box')

# Ajustar etiquetas del eje x
plt.xlabel('Tipo de Animal')
plt.xticks([0, 1], ['Perros', 'Gatos'])

# Agregar título al gráfico
plt.title('Boxplots del número promedio de perros y gatos')
```
